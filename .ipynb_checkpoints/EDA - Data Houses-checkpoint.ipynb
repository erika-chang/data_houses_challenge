## Baseline pipe ü™à

### Preprocessing

The preprocessing will be made differently depending on the type of the features:

Categorical features:

- Simple-impute with the most frequent values
- One-Hot Encode features that have less than 7 unique values to start with
- Drop all other features

Numerical features:

- Simple-Impute with strategy 'mean'
- Min-Max Scale

```python
preproc_numerical_baseline = make_pipeline(
    SimpleImputer(), 
    MinMaxScaler()
)
```

```python
preproc_categorical_baseline = make_pipeline(
    SimpleImputer(strategy='most_frequent'),
    OneHotEncoder(handle_unknown='ignore', drop='if_binary')
)
```

```python
preproc_baseline = make_column_transformer(
    (preproc_numerical_baseline, make_column_selector(dtype_include=['int64', 'float64'])),
    (preproc_categorical_baseline, feat_categorical_small),
    remainder='drop'
)
```

```python
preproc_baseline
```

```python
shape_preproc_baseline = preproc_baseline.fit_transform(X).shape
shape_preproc_baseline
```

### Estimator

```python
pipe_baseline = make_pipeline(preproc_baseline, DecisionTreeRegressor())
pipe_baseline
```

### Cross-validate

For this challenge, it is required that the evaluation method is the 'rmsle' (root mean squared log error), so that the lower the error is, the better the model used.

$$\text{RMSLE}(y, \hat{y}) = \sqrt{\frac{1}{n_\text{samples}} \sum_{i=0}^{n_\text{samples} - 1} (\log_e (1 + y_i) - \log_e (1 + \hat{y}_i))^2.}$$

```python
def root_mean_squared_log_error(y, y_pred):
    t = np.array(y)
    p = np.array(y_pred)
    
    log_error = np.log(1-t) - np.log()
    
    return ((log_error**2).mean())**0.5
```

```python
rmsle = make_scorer(root_mean_squared_log_error)
```

```python
cross_val_score(pipe_baseline, X, y, cv=5, scoring=rmsle)
```
